```python
import matplotlib.pyplot as plt
import pandas as pd
ai_file_path = 'ai_dataset.csv'
ai_df = pd.read_csv(ai_file_path)

#Display the first few rows of the AI dataset to see its structure
ai_df.head()

# Filtering data for jobs with high AI adoption and automation risk
high_ai_jobs = ai_df[ai_df['AI Adoption Level'] == 'High']
high_risk_jobs = high_ai_jobs[high_ai_jobs['Automation Risk'] == 'High']


# Objective 1: Analyze AI adoption by industry without summing or averaging duplicate values
# We will count the number of high AI adoption jobs in each industry instead of averaging scores
ai_adoption_by_industry = high_ai_jobs['Industry'].value_counts()

# Plotting the number of high AI adoption jobs by industry
plt.figure(figsize=(10,6))
ai_adoption_by_industry.plot(kind='bar', color='skyblue')
plt.title('Number of High AI Adoption Jobs by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Jobs with High AI Adoption')
plt.xticks(rotation=45)
plt.show()

# Objective 2: Identify top skills in high AI adoption industries
skills = high_ai_jobs['Required_Skills'].value_counts().head(10)

# Plotting the top 10 required skills
plt.figure(figsize=(10,6))
skills.plot(kind='barh', color='green')
plt.title('Top 10 Required Skills in High AI Adoption Industries')
plt.xlabel('Frequency')
plt.ylabel('Skills')
plt.show()

# Objective 3: Analyze job growth projections in high automation risk roles
growth_by_risk = high_risk_jobs.groupby('Job Growth Projection')['Job_Title'].count()

# Plotting job growth projections for high automation risk roles
plt.figure(figsize=(10,6))
growth_by_risk.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'gold', 'lightgreen'], startangle=140)
plt.title('Job Growth Projections for High Automation Risk Roles')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Objective 1: Analyze Job Displacement Trends in Industries with Significant AI Adoption ---

# Group by industry and calculate the percentage of high AI adoption jobs per industry
industry_job_counts = ai_df['Industry'].value_counts()
high_ai_job_counts = high_ai_jobs['Industry'].value_counts()

# Calculate percentage of jobs with high AI adoption in each industry
industry_high_ai_percentage = (high_ai_job_counts / industry_job_counts) * 100

# Sort by highest percentage
industry_high_ai_percentage = industry_high_ai_percentage.sort_values(ascending=False)

# Plot the percentage of high AI adoption jobs per industry
plt.figure(figsize=(10,6))
industry_high_ai_percentage.plot(kind='bar', color='teal')
plt.title('Percentage of Jobs with High AI Adoption by Industry')
plt.xlabel('Industry')
plt.ylabel('Percentage of High AI Adoption Jobs')
plt.xticks(rotation=45)
plt.show()

# --- Objective 2: Identify Obsolete Skills and New Competencies Required in an AI-Driven Workforce ---

# Calculate the frequency of each skill
skills = high_ai_jobs['Required_Skills'].value_counts()

# Calculate percentage of each skill
skills_percentage = (skills / skills.sum()) * 100

# Plot the top 10 skills as a pie chart
plt.figure(figsize=(8,8))
skills_percentage.head(10).plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Top 10 Required Skills in High AI Adoption Industries (Percentage)')
plt.ylabel('')
plt.show()

# --- Objective 3: Examine Societal Challenges Like Inequality and Job Displacement Due to AI ---

# Create a "Risk Factor" based on AI adoption score and automation risk score
high_risk_jobs['Risk_Factor'] = high_risk_jobs['AI Adoption Score'] * high_risk_jobs['Automation Risk Score']

# Analyze risk factor by job growth projections
risk_by_growth = high_risk_jobs.groupby('Job Growth Projection')['Risk_Factor'].mean()

# Plot the average risk factor by job growth projection
plt.figure(figsize=(10,6))
risk_by_growth.plot(kind='bar', color='coral')
plt.title('Average Risk Factor by Job Growth Projection')
plt.xlabel('Job Growth Projection')
plt.ylabel('Average Risk Factor')
plt.xticks(rotation=0)
plt.show()

# Calculate and display the overall mean risk factor for high AI adoption jobs
mean_risk_factor = high_risk_jobs['Risk_Factor'].mean()
print(f"Overall Mean Risk Factor for High AI Adoption Jobs: {mean_risk_factor:.2f}")
```
