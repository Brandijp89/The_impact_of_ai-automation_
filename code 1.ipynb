```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from collections import Counter

'''nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')'''



# Sample dataset of job descriptions
job_descriptions = ["AI Specialist required with Python skills",
                    "Looking for automation engineer",
                    "Data Scientist with machine learning expertise"]

# Tokenize and count keywords
stop_words = set(stopwords.words('english'))
words = [word for description in job_descriptions for word in nltk.word_tokenize(description.lower())
         if word.isalnum() and word not in stop_words]

word_freq = Counter(words)

print("Most common skills:", word_freq.most_common(5))
```

```python

import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Generate word cloud for skills
text = " ".join(word for word in words)
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data for job loss trend (year vs. job count)
years = np.array([2015, 2016, 2017, 2018, 2019]).reshape(-1, 1)
job_count = np.array([1000, 950, 900, 850, 800])  # Example job count data

# Fit a linear regression model
model = LinearRegression()
model.fit(years, job_count)

# Predict job count for the next 5 years
future_years = np.array([2020, 2021, 2022, 2023, 2024]).reshape(-1, 1)
predictions = model.predict(future_years)

print("Job loss predictions for future years:", predictions)
```
